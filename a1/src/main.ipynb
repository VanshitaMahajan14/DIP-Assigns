### Function to convert RGB image to grayscale. 
#### Here image[:,:,k] parses image into pixels. 
#### image[:,:,0] is for red pixel, denoted by r
#### image[:,:,1] is for green pixel, denoted by g
#### image[:,:,2] is for blue pixel, denoted by b

```python
def convert_rgb_to_grayscale(image):
    r, g, b = image[:,:,0], image[:,:,1], image[:,:,2]
    grayscale_image = 0.2989 * r + 0.5870 * g + 0.1140 * b
    grayscale_image = grayscale_image.astype(np.uint8)
    return grayscale_image
```

### converting all images to grayscale using the "convert_rgb_to_grayscale" function

```python
im1_gray = convert_rgb_to_grayscale(im1_rgb)  
im2_gray = convert_rgb_to_grayscale(im2_rgb)  
im3_gray = convert_rgb_to_grayscale(im3_rgb)  
im4_gray = convert_rgb_to_grayscale(im4_rgb)  

#display the grayscale images

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(im1_gray)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(im2_gray)
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(im3_gray)
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(im4_gray)
plt.axis('off')

plt.tight_layout()
plt.axis('off')
plt.show()
```

### saving all grayscale pictures in the "/images/opencv" folder

## USING PIL

```python
import os
from PIL import Image
```

```python
#loading images
images_folder = "../images"
image_file1 = "pic1.jpg"
image_file2 = "pic2.jpg"
image_file3 = "pic3.jpg"
image_file4 = "pic4.jpg"

pil_path1 = os.path.join(images_folder, image_file1)      #defining path
pil_image1 = Image.open(pil_path1)                        #opening image

pil_path2 = os.path.join(images_folder, image_file2)
pil_image2 = Image.open(pil_path2)

pil_path3 = os.path.join(images_folder, image_file3)
pil_image3 = Image.open(pil_path3)

pil_path4 = os.path.join(images_folder, image_file4)
pil_image4 = Image.open(pil_path4)

# Displaying the loaded images

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(pil_image1)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(pil_image2)
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(pil_image3)
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(pil_image4)
plt.axis('off')

plt.tight_layout()
plt.axis('off')
plt.show()
```

### converting loaded images to grayscale using in-built function in PIL

```python
pil_gray1 = pil_image1.convert('L')
pil_gray2 = pil_image2.convert('L')
pil_gray3 = pil_image3.convert('L')
pil_gray4 = pil_image4.convert('L')

#displaying converted grayscale images

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(pil_gray1)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(pil_gray2)
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(pil_gray3)
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(pil_gray4)
plt.axis('off')

plt.tight_layout()
plt.axis('off')
plt.show()
```

### saving converted grayscale images in "../images/pil" folder

```python
out_path1 = '../images/pil/gray_pic1.jpg'   #defining path
pil_gray1.save(out_path1)                #writing output image to path

out_path2 = '../images/pil/gray_pic2.jpg'
pil_gray2.save(out_path2)

out_path3 = '../images/pil/gray_pic3.jpg'
pil_gray3.save(out_path3)

out_path4 = '../images/pil/gray_pic4.jpg'
pil_gray4.save(out_path4)
```
